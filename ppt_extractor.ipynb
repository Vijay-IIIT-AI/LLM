```python
from unsloth import FastVisionModel
import torch
from PIL import Image
from tqdm import tqdm

# Global variable for the vision model and tokenizer
vision_model = None
vision_tokenizer = None

def load_vision_model(model_name: str, load_in_4bit: bool = True):
    """
    Loads the specified vision model and tokenizer globally.

    Args:
        model_name (str): The name of the model to load (e.g., "unsloth/Llama-3.2-11B-Vision-Instruct" or "unsloth/Pixtral-12B-Base-2409-bnb-4bit").
        load_in_4bit (bool): Whether to load the model in 4-bit precision (default: True). If False, loads in 16-bit.
    """
    global vision_model, vision_tokenizer

    vision_model, vision_tokenizer = FastVisionModel.from_pretrained(
        model_name,
        load_in_4bit=load_in_4bit,
        device_map="auto"
    )

    # Enable inference mode
    FastVisionModel.for_inference(vision_model)
    print(f"Model '{model_name}' loaded successfully with {'4-bit' if load_in_4bit else '16-bit'} precision.")

def get_vision_model():
    """
    Returns the loaded vision model and tokenizer.
    """
    if vision_model is None or vision_tokenizer is None:
        raise ValueError("Vision model is not loaded. Please call load_vision_model() first.")
    return vision_model, vision_tokenizer
```

```python
def generate_prompt(image_base, instruction):

    messages = [
        {"role": "user", "content": [
            {"type": "image"},
            {"type": "text", "text": instruction}
        ]}
    ]

    # Convert messages into a format the model understands
    input_text = vision_tokenizer.apply_chat_template(messages, add_generation_prompt=True)

    # Tokenize the input (text + image)
    inputs = vision_tokenizer(
        images=[image_base],  # Pass image as a list
        text=input_text,  # Ensure the text references the image
        add_special_tokens=False,
        return_tensors="pt"
    ).to("cuda")

    output_ids = vision_model.generate(**inputs, max_new_tokens=300, use_cache=True, temperature=1.0)
    # Decode the output to get plain text
    response_text = vision_tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return response_text
```

```python
def process_image(image_file):
  image_path = image_file # Replace with your image path
  image = Image.open(image_path).convert("RGB")  # Ensure image is in RGB format
  return image
```

```python
import os
import platform
import subprocess
import pandas as pd
from PIL import Image
from tqdm import tqdm

def get_system_type():
    return platform.system().lower()

def check_office_availability():
    system_type = get_system_type()
    if system_type == "windows":
        try:
            import win32com.client
            return True
        except ImportError:
            return False
    elif system_type == "linux":
        try:
            subprocess.run(["libreoffice", "--version"], stdout=subprocess.PIPE, stderr=subprocess.PIPE, check=True)
            return True
        except FileNotFoundError:
            return False
    return False

def convert_ppt_to_pdf_linux(pptfile_name, out_dir):
    pdf_path = os.path.join(out_dir, os.path.splitext(os.path.basename(pptfile_name))[0] + ".pdf")
    command = ["soffice", "--headless", "--convert-to", "pdf", "--outdir", out_dir, pptfile_name]
    subprocess.run(command, check=True)
    if not os.path.exists(pdf_path):
        raise RuntimeError("PDF conversion failed.")
    return pdf_path

def convert_pdf_to_images_linux(pdf_path, out_dir, img_format="png"):
    from pdf2image import convert_from_path
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)
    images = convert_from_path(pdf_path, dpi=300, fmt=img_format)
    image_paths = []
    for i, image in enumerate(images):
        img_path = os.path.join(out_dir, f"slide_{i+1}.{img_format}")
        image.save(img_path, img_format.upper())
        image_paths.append(img_path)
    if not image_paths:
        raise RuntimeError("Image extraction failed.")
    return image_paths

def convert_ppt_to_images_windows(pptfile_name, out_dir, img_format="jpg"):
    import win32com.client
    img_format_map = {"jpg": 17, "jpeg": 17, "png": 18, "bmp": 19}

    if img_format.lower() not in img_format_map:
        raise ValueError(f"Unsupported format: {img_format}. Use 'jpg', 'png', or 'bmp'.")

    powerpoint = win32com.client.Dispatch("PowerPoint.Application")
    powerpoint.Visible = 1
    presentation = powerpoint.Presentations.Open(os.path.abspath(pptfile_name), WithWindow=0)
    presentation.SaveAs(os.path.abspath(out_dir), img_format_map[img_format.lower()])

    slide_count = presentation.Slides.Count
    presentation.Close()
    powerpoint.Quit()

    images = [os.path.join(out_dir, f"Slide{i+1}.{img_format}") for i in range(slide_count)]
    print(f"Slides saved in: {out_dir}")
    return images

def extract_ppt_slides(pptx_path, output_folder):
    if not pptx_path.endswith(".pptx"):
        raise ValueError("Invalid file format. Only .pptx files are allowed.")

    system_type = get_system_type()
    pptx_filename = os.path.basename(pptx_path).split('.')[0]
    ppt_output_folder = os.path.join(output_folder, f"{pptx_filename}_ppt_image_extracted")
    os.makedirs(ppt_output_folder, exist_ok=True)
    slides_folder = os.path.join(ppt_output_folder, "slides")
    os.makedirs(slides_folder, exist_ok=True)
    csv_data = []
    pdf_path = None

    if check_office_availability():
        print("Office software available, using system extraction.")
        if system_type == "windows":
            images = convert_ppt_to_images_windows(pptx_path, slides_folder)
        else:
            pdf_path = convert_ppt_to_pdf_linux(pptx_path, ppt_output_folder)
            images = convert_pdf_to_images_linux(pdf_path, slides_folder)
    else:
        raise RuntimeError("Windows Office or LibreOffice not found! Cannot process PPTX file.")

    for i, img_path in enumerate(images):
        csv_data.append([i+1, pptx_filename, img_path])

    csv_path = os.path.join(ppt_output_folder, "slide_mapping.csv")
    df = pd.DataFrame(csv_data, columns=["Slide Number", "PPT File Name", "Image File Path"])
    df.to_csv(csv_path, index=False)
    print(f"Slides extracted successfully to {ppt_output_folder}")

    return images, csv_path, pdf_path
```

```python
prompt_image = """
    Extract all texts from this presentation slide (literal extraction, do not summarize!) and tabular data, when image or diagram is present describe it.
    When a graph with data is present, describe what it represents in detail, with key messages on data points.  When extracting texts, those that are part of a diagram or graph should be only extracted within this context, not within the "text" category.
    Don't miss any piece of information

    Write the extracted contents in JSON format, that will follow this structure:

    [
        {
            "heading": "Example slide heading",
            "text": "Example text on the slide"
            "images": [
                "Example image or diagram description",
                "Example image or diagram description",
                "Example image or diagram description"
            ],
            "graphs": [
                "Example graph description"
            ]
        }
    ]

    If there is any part missing (like there is no heading on the slide) - just represent it as null.
    Output the JSON file. Make sure your response is a VALID JSON! Do not put backticks around the output.

    SYNTACTICALLY CORRECT EXAMPLE RESPONSE:
    [
        {
            "slide_heading": "Ecological Threat Report 2023",
            "subheading": null,
            "text": "Key Findings\nCountry Hotspots\nEcological Threats\nMegacities\nPolicy Recommendations",
            "images": [
                "Example image"
            ],
            "graphs": []
        }
    ]
    """
```

```python
retrieval_data = []

# Wrap the iterrows with tqdm for progress tracking
for _, row in tqdm(data_mapping.iterrows(), total=data_mapping.shape[0], desc="Processing Images"):
    print("processing",_)
    image_path = row["Image File Path"]

    if os.path.exists(image_path):
        processed_image_value = process_image(image_path)
        result = generate_prompt(processed_image_value, prompt_image)
    else:
        result = "Image not found"

    retrieval_data.append(result)

data_mapping["Retrieval Data"] = retrieval_data
```
